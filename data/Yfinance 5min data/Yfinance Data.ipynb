```python
import yfinance as yf
from tensorflow import keras
import pandas as pd
import numpy as np
```

```python
# replace it with the path of IWB_holdings.csv on your device
path = '/Users/kalo/Downloads/IWB_holdings.csv'
```

```python
data = pd.read_csv(path,skiprows = 9)
tickernames = data['Ticker'].values[:-2]
tickernames = np.delete(tickernames,-3)
```

```python
tickernames
```

```python
# start_date = '2010-01-01'
# end_date = '2020-12-31'

# Create an empty DataFrame to store the data
intraday_data = []

# Loop through the tickers and fetch intraday data
for ticker in tickernames:
    # Download intraday data at a 5-minute interval within last 60 days
    data = yf.download(ticker, period = '60d', interval="5m")
    
    # Rename columns to include the ticker symbol
    data.columns = [f"{ticker}_{col}" for col in data.columns]
    
    # Concatenate data with the intraday_data DataFrame
    # Ignore tickers with missing data
    if len(data)==4680:
        intraday_data.append(data)

# Save the data to a CSV file or perform further analysis
# intraday_data.to_csv('russell_1000_intraday_data.csv')
```

```python
count = 0
for i in range(len(intraday_data)):
    if(len(intraday_data[i])<4680):
        count+=1
print('There are total '+str(len(tickernames))+' tickers')
print('There are '+str(len(intraday_data))+' ticker downloaded successfully')
```
